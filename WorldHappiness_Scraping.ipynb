```python
from selenium.webdriver.common.by import By
import pandas as pd
import time

all_data = []
years = ['2024', '2023', '2022', '2021', '2020', '2019', '2018', '2017', '2016', '2015', '2014']  # or dynamic later

for year in years:
    # Click dropdown
    driver.find_element(By.CSS_SELECTOR, 'div.selectedOption').click()
    time.sleep(0.2)  # tiny pause to open dropdown

    # Click year option
    driver.find_element(By.CSS_SELECTOR, f'div.option_{year}').click()
    time.sleep(0.5)  # short pause for table to render

    # Scrape table rows
    rows = driver.find_elements(By.CSS_SELECTOR, 'div[role="row"]')
    data = [[cell.text for cell in row.find_elements(By.CSS_SELECTOR, 'div[role="gridcell"]')]
            for row in rows[1:]]  # skip header

    # Add year
    data = [[year] + row for row in data]
    all_data.extend(data)
```

```python
columns_with_year = ['Year'] + columns
df = pd.DataFrame(all_data, columns=columns_with_year)
```
